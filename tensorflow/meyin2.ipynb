```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
a = np.array([10,21,31,41,51,61,71,81,91])
b = np.array([10,20,30,40,50,60,70,80,90])
```

```python
plt.plot(a,b,"g")
plt.xlabel("Ahmet")
plt.ylabel("Mehmet")
plt.title("Kazanma")
```

```python
c = np.linspace(0,10,5)
```

```python
c
```

```python
d = c ** 3
```

```python
plt.plot(c,d,"g*-")
```

```python
plt.subplot(1,2,1) # 1 sra 2 kolon 1. grafik
plt.plot(a,b,"r*-")
plt.subplot(1,2,2)
plt.plot(c,d,"g--")
```

```python
benimfigur = plt.figure()
figureAxes = benimfigur.add_axes([0.2,0.2,0.6,0.6]) # . . genilik ykseklik
figureAxes.plot(c,d,"r--")
figureAxes.set_xlabel("Ahmet")
figureAxes.set_ylabel("Mehmet")
figureAxes.set_title("Mahmut")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9]) # y'ye uzaklk x'e uzaklk genilik ykseklik
eksen2 = figure2.add_axes([0.3,0.7,0.2,0.2])

eksen1.plot(a,b,"g")
eksen1.set_xlabel("Ahmet1")
eksen1.set_ylabel("Mehmet1")
eksen1.set_title("Mahmut1")

eksen2.plot(c,d,"b")
eksen2.set_xlabel("Ahmet2")
eksen2.set_ylabel("Mehmet2")
eksen2.set_title("Mahmut2")
```

```python
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(a,b,"c")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)

for i in benimEksenler:
    i.plot(a,b,"b")
    i.set_xlabel("X ekseni")
    
plt.tight_layout
```

```python
newfigur = plt.figure()
# dpi=150

neweksen = newfigur.add_axes([0.1,0.1,0.9,0.9])
neweksen.plot(a, b, label = "b ** 2")
neweksen.plot(c, d, label = "a ** 3")
neweksen.legend()
#legend loc=2
```

```python
newfigur.savefig("benimfigur.png", dpi=200)
```
