```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_Marks.csv")
```

```python
df.head()
```

```python
y = df["Marks"].values
x = df.drop("Marks",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=19)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(2,activation="relu"))
model.add(Dense(2,activation="relu"))
model.add(Dense(1,activation="relu"))
model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
earstp = EarlyStopping(monitor="val_loss",mode="min", patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test), callbacks=[earstp])
```

```python
kayip = pd.DataFrame(model.history.history)
```

```python
kayip.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
tahminx = model.predict(x_test)
```

```python
mean_absolute_error(y_test,tahminx)
```

```python
plt.scatter(y_test,tahminx)
plt.plot(y_test,y_test,"g*-")
```
