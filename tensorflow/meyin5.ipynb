```python
y = df["Type"].values
x = df.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=19)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
#sigmoid 0 ve 1 arasında değer döner
#sınıflandırma işlemi yapıyoruz 0 ve 1
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test), verbose=0)
```

```python
kayip = pd.DataFrame(model.history.history)
```

```python
kayip.plot()
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
earlystopping = EarlyStopping(monitor="val_loss", mode="min", patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test), verbose=1, callbacks=[earlystopping])
```

```python
kayip2 = pd.DataFrame(model.history.history)
```

```python
kayip2.plot()
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test), verbose=1, callbacks=[earlystopping])
```

```python
kayip3 = pd.DataFrame(model.history.history)
```

```python
kayip3.plot()
```
